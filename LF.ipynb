```python
df = pd.read_csv("model_K_12to20mag.csv")
df_2 = pd.read_csv("model_K_12to18mag.csv")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator


import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator

def histlineplot(df, color, label):
    min_bin = min(df['K'].min(), 10)  # Asegurarse de incluir 10 en el rango de bins
    max_bin = df['K'].max()  # Asegurarse de incluir 20 en el rango de bins

    bins = np.arange(min_bin, max_bin + 0.5, 0.5)
    counts, edges = np.histogram(df['K'], bins=bins)
    bin_centers = 0.5 * (edges[:-1] + edges[1:])

    plt.plot(bin_centers, counts, marker='o', linestyle='-', color=color, label=label)
    plt.xlabel('K[mag]')
    plt.ylim(-50, 350)
    plt.xlim(10, 20)  # Configuración del rango x mostrado en el plot

    # Configuración de los ticks en el eje x
    ax = plt.gca()  # Obtener la instancia actual de Axes
    ax.xaxis.set_major_locator(MultipleLocator(2))
    ax.xaxis.set_minor_locator(MultipleLocator(0.5))

    # Configuración de los ticks en el eje y cada 100 unidades y cada 10 unidades para los minor ticks
    ax.yaxis.set_major_locator(MultipleLocator(100))
    ax.yaxis.set_minor_locator(MultipleLocator(10))

    plt.legend()
  #  plt.show()
```

```python
histlineplot(df,"orange","12<K<20")
histlineplot(df_2,"red","12<K<18")
plt.legend()
```
